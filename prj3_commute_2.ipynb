## 하차기준으로 데이터 분석

* 1.출근시간 07~09 시 사이에 하차 하는 역별 이용객수 정보 (회사)
* 4.퇴근시간 18-20 시 사이에 하차 하는 역별 이용객수 정보 (주택가)

```python
#출근시간에 하차가 가장 많은 역 (회사가 많은 역)
#퇴근시간에 하차가 가장 많은 역 (주택가)
df_getoff = df_m.loc[df_m['구분']=='하차',['년월일','역번호','역명','7시','8시','18시','19시','역위도','역경도']]
df_getoff.head()
```

### 데이터를 가져왔으니 이제 분석을 해볼까? 1,2번은 회사 3,4번이 주택가로 추정되는 역이다. 

```python
#1.하차정보로만 확인해보기 
df_getoff.head()
```

```python
#출근시간 7시8시 합해서 아침이란 컬럼으로 새로생성
#퇴근시간 18시19시 합해서 저녁이란 컬럼으로 새로 생성
df_getoff['아침'] = df_getoff['7시']+df_getoff['8시']
df_getoff['저녁'] = df_getoff['18시']+df_getoff['19시']
```

```python
df_getoff.head()
```

```python
#주중 평균으로 해야 조금더 정확할것 같다 토요일 일요일 제외하고 아침, 저녁 평균 이용객수를 산출하자
#주를 구하려면 년월일 컬럼을 date 타입으로 변경이 우선 필요하다.
df_getoff['년월일']=df_getoff['년월일'].astype('str')  # int -> str변환후 날짜 변환해야 함
df_getoff['년월일']=pd.to_datetime(df_getoff['년월일'])
df_getoff.info()
```

```python
#요일 반영하기
df_getoff['요일']= df_getoff['년월일'].dt.weekday.replace(0,'월').replace(1,'화').replace(2,'수').replace(3,'목').replace(4,'금').replace(5,'토').replace(6,'일')
df_getoff.tail()
```

```python
#평일만 가져와서 데이터프레임 재구성해보자
df_getoff=df_getoff[df_getoff['요일'].isin(['월','화','수','목','금'])]
df_getoff.head()
```

```python
#group by로  일평균 아침 / 저녁 하차 수를 역별로 계산해보자
dfg_getoff_day= df_getoff.groupby(['역번호','역명','역위도','역경도'])['아침','저녁'].mean()
dfg_getoff_day
```

```python
#인덱스 리셋
dfg_getoff_day = dfg_getoff_day.reset_index()
dfg_getoff_day.head()
```

```python
#아침 하차상위 top 30 
dfg_getoff_day=dfg_getoff_day.sort_values(by='아침',ascending=False)
getoff_am_top=dfg_getoff_day.head(30)
getoff_am_top.head()
```

```python
getoff_am_top = getoff_am_top.reset_index()

getoff_am_top.head()
```

```python
del getoff_am_top['index']
getoff_am_top.head()
```

```python
#저녁 하차상위 top 30 
dfg_getoff_day=dfg_getoff_day.sort_values(by='저녁',ascending=False)
getoff_pm_top=dfg_getoff_day.head(30)
```

```python
getoff_pm_top = getoff_pm_top.reset_index()
del getoff_pm_top['index']
getoff_pm_top.head()
```

```python
#지도 시각화 
import folium
map_osm = folium.Map(location=[35.166804,129.083479], zoom_start=12)   # 지도 초기 로딩위치
```

```python
getoff_am_top.index
```

```python
#지도로 출근시간(아침) 이용객 30개 역 표시
for item in getoff_am_top.index:
    lat = getoff_am_top.loc[item,'역위도']
    long = getoff_am_top.loc[item,'역경도']
    folium.CircleMarker([lat,long],  #위도/경도
                        radius=getoff_am_top.loc[item,'아침']/200,
                        popup=getoff_am_top.loc[item,'역명'],
                        color ='blue',
                        fill = True).add_to(map_osm)

map_osm
```

```python
#지도로 퇴근시간(저녁) 상위 30개 이용객 표시 
for item in getoff_pm_top.index:
    lat = getoff_pm_top.loc[item,'역위도']
    long = getoff_pm_top.loc[item,'역경도']
    folium.CircleMarker([lat,long],  #위도/경도
                        radius=getoff_pm_top.loc[item,'저녁']/200,
                        popup=getoff_pm_top.loc[item,'역명'],
                        color ='red',
                        fill = True).add_to(map_osm)
map_osm
```
